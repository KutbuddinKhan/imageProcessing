**Libraries that can be used for image processing**
1. matplotlib.image
2. pillow
3. OpenCV(cv2)

```python
# Importing the image module from patplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Loading an image through matplotlib.image module
img = mpimg.imread('puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
print(img)
```

```python
# Displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

**Resizing the image using pillow Library**

```python
from PIL import Image
```

```python
img = Image.open('puppy.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('puppy_image_resize.jpg')
```

```python
# Displaying the image from numpy array

img_res = mpimg.imread('puppy_image_resize.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
img_res.shape
```

**Convert RGB images to GrayScale image using OpenCV**

```python
# importing openCV Library
import cv2
```

```python
img = cv2.imread('puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image. But this will not be allowed in google colab
